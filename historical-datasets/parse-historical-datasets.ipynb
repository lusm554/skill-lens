```python
def preview_ds(filepath, sep=','):
    print(filepath)
    !head -n2 {filepath}
    print()
    df = pl.scan_csv(filepath, separator=sep, encoding='utf8-lossy', ignore_errors=True, infer_schema_length=0)
    print(df.collect_schema().names())
    return df.head(3).collect()
```
