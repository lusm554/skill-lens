# Join datasets

```python
columns_mapping = {
    'vacancy_id': {
        'data_analyst_vacancies_hh_by_lludop.csv': 'id_vac',
        'it_vacancies_hh_by_ilyazawilsiv.csv': 'id',
        'it_vacancy_data_hh_by_figshare.csv': 'id',
        'jobs_raw_data_hh_by_etietopabraham.csv': 'id',
        # 'resume_job_matching_dataset_hh_by_darysha.csv': '',
        'vacancies_data_hh_trudvsem_superjob_rabotaru_by_mendeley_data.csv': 'id',
    },
    'vacancy_name': {
        'data_analyst_vacancies_hh_by_lludop.csv': 'name',
        'it_vacancies_hh_by_ilyazawilsiv.csv': 'name',
        'it_vacancy_data_hh_by_figshare.csv': 'name',
        'jobs_raw_data_hh_by_etietopabraham.csv': 'title',
        'resume_job_matching_dataset_hh_by_darysha.csv': 'positionName',
        'vacancies_data_hh_trudvsem_superjob_rabotaru_by_mendeley_data.csv': '',
    },
    'date_published': {
        'data_analyst_vacancies_hh_by_lludop.csv': 'published_at',
        'it_vacancies_hh_by_ilyazawilsiv.csv': 'created_at',
        'it_vacancy_data_hh_by_figshare.csv': 'published_at',
        'jobs_raw_data_hh_by_etietopabraham.csv': 'date_of_post',
        'resume_job_matching_dataset_hh_by_darysha.csv': 'publishedDate',
        'vacancies_data_hh_trudvsem_superjob_rabotaru_by_mendeley_data.csv': 'last_found_at',
    },
    'salary_from': {
        'data_analyst_vacancies_hh_by_lludop.csv': 'salary',
        'it_vacancies_hh_by_ilyazawilsiv.csv': 'salary_from',
        # 'it_vacancy_data_hh_by_figshare.csv': '',
        'jobs_raw_data_hh_by_etietopabraham.csv': 'salary',
        # 'resume_job_matching_dataset_hh_by_darysha.csv': '',
        'vacancies_data_hh_trudvsem_superjob_rabotaru_by_mendeley_data.csv': 'salary_from',
    },
    'data_source': {
        'vacancies_data_hh_trudvsem_superjob_rabotaru_by_mendeley_data.csv': 'data_source',
    }
}
columns_by_file = {
    f: {
        key: mapp.get(f)
        for key, mapp in columns_mapping.items()
        if mapp.get(f)
    }
    for f in [Path(fpath).name for fpath in datasets_filepath_list]
}
col_types_mapping = {
    'vacancy_id': pl.String,
    'vacancy_name': pl.String,
    # 'salary_from': pl.Float64,
    'salary_from': pl.String,
    'data_source': pl.String,
}
dt_formats_mapping = {
    "data_analyst_vacancies_hh_by_lludop.csv": "%d.%m.%Y %H:%M",
    "it_vacancies_hh_by_ilyazawilsiv.csv": "%Y-%m-%d %H:%M:%S%z",
    "it_vacancy_data_hh_by_figshare.csv": "%Y-%m-%d",
    "jobs_raw_data_hh_by_etietopabraham.csv": "%d.%m.%Y",
    "resume_job_matching_dataset_hh_by_darysha.csv": "%Y-%m-%d %H:%M:%S%z",
    "vacancies_data_hh_trudvsem_superjob_rabotaru_by_mendeley_data.csv": "%Y-%m-%d %H:%M:%S",
}
```

```python
# Московский часовой пояс (pytz нужен для utcOffset)
MOSCOW_TZ = "Europe/Moscow"

def date_expr(col, fmt, has_tz):
    expr = pl.col(col).cast(pl.Utf8).str.strptime(
        pl.Datetime, fmt, strict=False
    )
    if has_tz:
        # Если уже есть timezone, просто приводим к московской
        expr = expr.dt.convert_time_zone(MOSCOW_TZ)
    else:
        # Для naive — добавляем московскую тк как локальную
        expr = expr.dt.replace_time_zone(MOSCOW_TZ)
    # Всегда сериализуем к ISO формату c таймзоной
    expr = expr.dt.strftime("%Y-%m-%dT%H:%M:%S%z")
    return expr
```

```python
lazy_datasets = list()
for filepath in datasets_filepath_list:
    filename = Path(filepath).name
    file_sep = sep_mapping.get(filename)
    dt_fmt = dt_formats_mapping.get(filename)
    has_tz = "%z" in dt_fmt
    df = pl.scan_csv(filepath, separator=file_sep, encoding='utf8-lossy', ignore_errors=True, infer_schema_length=0)
    cols_mapping = columns_by_file.get(filename)
    cols_mapping_reverse = dict(zip(cols_mapping.values(), cols_mapping.keys()))
    col_types = [pl.col(c).cast(col_types_mapping.get(c)) for c in cols_mapping if c!='date_published']
    if 'date_published' in cols_mapping:
        col_types.append(date_expr('date_published', dt_fmt, has_tz).alias('date_published'))
    df = (
        df
            .select(cols_mapping.values())
            .rename(cols_mapping_reverse)
            .select(col_types)
            .with_columns(pl.lit(filename).alias('file_source'))
    )
    if 'data_source' not in df.collect_schema().names():
        df = df.with_columns(pl.lit('hh').alias('data_source'))
    lazy_datasets.append(df)
    print(filename)
    print(df.head().collect())
    print()
```

```python
lazy_all = pl.concat(lazy_datasets, how="diagonal")
lazy_all = lazy_all.drop_nulls()
```

```python
%%time
save_filepath = '../data/historical-datasets/all_datasets.parquet'
lazy_all.sink_parquet(save_filepath, compression_level=15)
```
